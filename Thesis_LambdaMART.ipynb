BM25 indexing

```python
%%writefile 01_full_indexing_bm25.py
from codecarbon import EmissionsTracker
import subprocess
import os
#setup tracker
tracker = EmissionsTracker(
    project_name="1_BM25_Indexing",
    save_to_file=True,
    output_file="emissions.csv",
    measure_power_secs=5,
    tracking_mode="process"
)

tracker.start()

try:
    os.makedirs("collections/msmarco-passage/collection_jsonl", exist_ok=True)

    # first convert tsv towards JSONL
    convert_cmd = [
        "python3.8", "tools/scripts/msmarco/convert_collection_to_jsonl.py",
        "--collection-path", "collections/msmarco-passage/collection.tsv",
        "--output-folder", "collections/msmarco-passage/collection_jsonl"
    ]
    subprocess.run(convert_cmd, check=True)

    # using pyserini lucene to do indexing
    bm25_indexing = (
        "python3.8 -m pyserini.index.lucene --collection JsonCollection "
        "--input collections/msmarco-passage/collection_jsonl "
        "--index indexes/bm25 --generator DefaultLuceneDocumentGenerator "
        "--threads 4 --storePositions --storeDocvectors --storeRaw"
    )
    subprocess.run(bm25_indexing, shell=True, check=True)

except subprocess.CalledProcessError as e:
    print(f"!! Error during execution: {e}")

finally:
    emissions = tracker.stop()
```

```python
%%writefile 02_retrieve_bm25.py
from codecarbon import EmissionsTracker
import subprocess

tracker = EmissionsTracker(
    project_name="2_BM25_Retrieval",
    save_to_file=True,
    output_file="emissions.csv",
    measure_power_secs=5,
    tracking_mode="process"
    )

tracker.start()
# retrieves the top 1000 results based on bm25, based on this link https://github.com/castorini/pyserini/blob/d4a4e390abad1c3185ff5121931454bfc383c2dd/docs/experiments-msmarco-passage.md
bm25_retrieval = (
        "python3.8", "-m", "pyserini.search.lucene",
        "--index", "indexes/bm25",
        "--topics", "dl19-passage",
        "--output", "runs/run.dl19.bm25.tsv",
        "--output-format", "msmarco",
        "--hits", "1000",
        "--bm25",
        "--k1", "0.82",
        "--b", "0.68",
        "--threads", "4",
        "--batch-size", "16"
)
subprocess.run(bm25_retrieval, check=True)
tracker.stop()
```

LambdaMART indexing

```python
%%writefile 03_full_indexing_ltr.py
from codecarbon import EmissionsTracker
import subprocess

tracker = EmissionsTracker(
    project_name="3_LambdaMART_Indexing",
    output_file="emissions.csv",
    measure_power_secs=5,
    tracking_mode="process"
)
tracker.start()

try:
    print("Retrieving top1000 pids")
    # pid = passage id
    top1000_pid = set()

    with open("runs/run.dl19.bm25.tsv", "r", encoding="utf-8") as run_file:
        for line in run_file:
            parts = line.strip().split("\t")
            # check if parts contain qid[0] and pid[1]
            if len(parts) >= 2:
                top1000_pid.add(parts[1].strip())

    print(f"Retrieved {len(top1000_pid)} pids from BM25 run.")

    full_collection = "collections/msmarco-passage/collection.tsv"
    filtered_collection = "collections/msmarco-passage/filtered_collection.tsv"

    matched_count = 0
    print("Retrieving passages from top1000 pids")

    # only select the top 1000 pids from the full MSMARCO collection
    with open(full_collection, "r", encoding="utf-8", errors="ignore") as input_file:
        with open(filtered_collection, "w", encoding="utf-8") as output_file:
            for line in input_file:
                parts = line.split("\t", 1)
                if parts:
                    current_pid = parts[0].strip()
                    if current_pid in top1000_pid:
                        output_file.write(line)
                        matched_count += 1
                        if matched_count % 10000 == 0:
                            print(f" {matched_count} matching passages...")

    print(f"Wrote {matched_count} passages to {filtered_collection}")


    print("Converting raw text into json format ")
    convert_cmd = [
        "python3.8", "scripts/ltr_msmarco/convert_passage.py",
        "--input", filtered_collection,
        "--output", "collections/msmarco-ltr-passage/ltr_collection.json",
        "--proc_qty", "4"
    ]
    subprocess.run(convert_cmd, check=True)


    print("JSON to JSONL")
    jsonl_cmd = [
        "python3.8", "scripts/ltr_msmarco/convert_collection_to_jsonl.py",
        "--collection-path", "collections/msmarco-ltr-passage/ltr_collection.json",
        "--output-folder", "collections/msmarco-ltr-passage/ltr_collection_jsonl"
    ]
    subprocess.run(jsonl_cmd, check=True)

    print("LTR indexing using Lucence")
    index_cmd = [
        "python3.8", "-m", "pyserini.index.lucene",
        "--collection", "JsonCollection",
        "--input", "collections/msmarco-ltr-passage/ltr_collection_jsonl",
        "--index", "indexes/msmarco-passage-ltr",
        "--generator", "DefaultLuceneDocumentGenerator",
        "--threads", "4",
        "--storePositions", "--storeDocvectors", "--storeRaw", "--pretokenized"
    ]
    subprocess.run(index_cmd, check=True)
    print(" Index built successfully.")

except Exception as e:
    print(f" Error {e}")
finally:
    tracker.stop()
```

LambdaMART reranking

```python
%%writefile 04_rerank_ltr.py
from codecarbon import EmissionsTracker
import subprocess
import os
import sys



tracker = EmissionsTracker(
    project_name="4_LambdaMART_Reranking",
    output_file="emissions.csv",
    measure_power_secs=5,
    tracking_mode="process",
)

tracker.start()

try:
    print("LambdaMART Reranking on filtered top 1000")
    rerank_cmd = [
        "python3.8", "pyserini/search/lucene/ltr/__main__.py",
        "--index", "indexes/msmarco-passage-ltr",
        "--model", "/content/drive/MyDrive/LamdaMART_MSMARCO_thesis/collections/msmarco-ltr-passage/msmarco-passage-ltr-mrr-v1",
        "--ibm-model", "/content/drive/MyDrive/LamdaMART_MSMARCO_thesis/collections/msmarco-ltr-passage/ibm_model/",
        "--topic", "tools/topics-and-qrels/topics.dl19-passage.txt",
        "--qrel", "tools/topics-and-qrels/qrels.dl19-passage.txt",
        "--input", "runs/run.dl19.bm25.tsv",
        "--input-format", "tsv",
        "--output", "/content/drive/MyDrive/LamdaMART_MSMARCO_thesis/runs/run.ltr.msmarco-passage.tsv",
        "--output-format", "tsv",
        "--hits", "100",
        "--granularity", "passage",
        "--rerank"
    ]


    env = os.environ.copy()
    env["PYTHONPATH"] = "."

    subprocess.run(rerank_cmd, capture_output=True, text=True, env=env)
    print("LambdaMART reranking complete.")

except subprocess.CalledProcessError as e:
    print(f"Error: {e}")

finally:
    tracker.stop()
```
